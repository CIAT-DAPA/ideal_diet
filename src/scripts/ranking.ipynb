```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import datetime
from scipy.spatial.distance import pdist, squareform
import math 


# Global variables
path_root = "G:\\CIAT\\Code\\CWR\\ideal_diet\\src\\"
path_inputs = path_root +"inputs\\"
path_outputs = path_root + "outputs\\"
path_results = path_outputs + "Results.xlsx"

# Getting data
data = pd.read_excel(path_results, sheet_name='ranking_countries')
print(data.head())

```

```python
import plotly
import plotly.plotly as py

df = {  'A':data.sort_values('OrderA')["Area"],
        'B':data.sort_values('OrderB')["Area"],
        'Total':data.sort_values('OrderTotal')["Area"],
        'Amount':data.Total}
df = pd.DataFrame(df)

labels = pd.concat([df.A, df.B])
#targets = pd.concat([, data.OrderB])
print(labels.to_string())
data_p = dict(
    type='sankey',
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label = labels,
      color = ["blue", "blue", "blue", "blue", "blue", "blue"]
    ),
    link = dict(
      source = data.OrderA,
      target = data.OrderB+217,
      value = data.Total
  ))
    
layout =  dict(
    title = "Basic Sankey Diagram",
    font = dict(
      size = 10
    )
)
       
fig = dict(data=[data_p], layout=layout)
    
#df = pd.DataFrame(A=data_A,B=data_B,Total=data_Total,amount=data.Total)
#print(data_A.head())
#print(data_B.head())
#fig = genSankey(df,cat_cols=['A','B'],value_cols='Amount',title='Ranking')
plotly.offline.plot(fig, validate=False)
```
